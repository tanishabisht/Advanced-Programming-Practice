**NOTE:** the database only stores values of the following fields: date, receipt number, rental company info (company, representative, location)

```python
from tkinter import *
from tkinter import ttk
import sqlite3

# Creating main page widget

main = Tk()
main.title('Set 5')
main.configure(background="#fff")





#################
### DATABASES ###
#################

# create or connect to a db
conn = sqlite3.connect('./Database/set5.db')

# create a cursor
c = conn.cursor()

# create a table
# c.execute("""CREATE TABLE carreceipt (
#     date text,
#     receiptNo integer,
#     company text,
#     representative text,
#     location text
# )""")

# create funmit function for database
def submit():
    conn = sqlite3.connect('./Database/set5.db')
    c = conn.cursor()
    
    # insert into table
    c.execute("INSERT INTO carreceipt VALUES (:date, :receiptNo, :company, :representative, :location)", {
        'date': e1.get(),
        'receiptNo': e2.get(),
        'company': Ecompany.get(),
        'representative': Erepresentative.get(),
        'location': Elocation.get()
    })
    
    c.execute("SELECT * FROM carreceipt")
    print('The values stored in the database are: ')
    print(c.fetchall())    
    
    conn.commit()
    conn.close()
    
    e1.delete(0, END)
    e2.delete(0, END)
    Ecompany.delete(0, END)
    Erepresentative.delete(0, END)
    Elocation.delete(0, END)
    
    





###############
### TKINTER ###
###############


root = LabelFrame(main, padx=40, pady=20)
root.pack(padx=30, pady=30)
root.configure(borderwidth=0, background="#fff")



Label(root, text='CAR RENTAL RECEIPT', bg='#fff', font='Arial 20').grid(sticky='W', row=0, column=0)
Label(root, text='', bg='#fff').grid(row=1)




# DATE
date = LabelFrame(root)
date.grid(sticky="W", row=2, column=0)
date.configure(borderwidth=0)
l1 = Label(date, text='Date: ', bg='#fff').pack(side=LEFT)
e1 = Entry(date)
e1.pack(side=LEFT)




# RECEIPT
reciept = LabelFrame(root)
reciept.grid(sticky="W", row=3, column=0)
reciept.configure(borderwidth=0)
l2 = Label(reciept, text='Receipt #: ', bg='#fff').pack(side=LEFT)
e2 = Entry(reciept)
e2.pack(side=LEFT)



# SPACE
Label(root, text='', bg='#fff').grid(row=4)




# RENTAL COMPANY INFO
rentalInfo = LabelFrame(root)
rentalInfo.grid(sticky="W", row=5, column=0)
rentalInfo.configure(borderwidth=0, bg='#fff')

Label(rentalInfo, text='Rental Company Info', bg='#fff', font='Arial 10 bold').grid(sticky='W')

company = LabelFrame(rentalInfo)
company.grid(sticky="W")
company.configure(borderwidth=0)
Label(company, text='Company: ', bg='#fff').pack(side=LEFT)
Ecompany = Entry(company)
Ecompany.pack(side=LEFT)

representative = LabelFrame(rentalInfo)
representative.grid(sticky="W")
representative.configure(borderwidth=0)
Label(representative, text='Representative: ', bg='#fff').pack(side=LEFT)
Erepresentative = Entry(representative)
Erepresentative.pack(side=LEFT)

location = LabelFrame(rentalInfo)
location.grid(sticky="W")
location.configure(borderwidth=0)
Label(location, text='Location: ', bg='#fff').pack(side=LEFT)
Elocation = Entry(location)
Elocation.pack(side=LEFT)

city = LabelFrame(rentalInfo)
city.grid(sticky="W")
city.configure(borderwidth=0)
Label(city, text='City/State/ZIP: ', bg='#fff').pack(side=LEFT)
Entry(city).pack(side=LEFT)

phone = LabelFrame(rentalInfo)
phone.grid(sticky="W")
phone.configure(borderwidth=0)
Label(phone, text='Phone: ', bg='#fff').pack(side=LEFT)
Entry(phone).pack(side=LEFT)




# LESSEE INFO
lesseeInfo = LabelFrame(root)
lesseeInfo.grid(sticky="W", row=5, column=1)
lesseeInfo.configure(borderwidth=0, bg='#fff')

Label(lesseeInfo, text='Lessee Info', bg='#fff', font='Arial 10 bold').grid(sticky='W')

company = LabelFrame(lesseeInfo)
company.grid(sticky="W")
company.configure(borderwidth=0)
Label(company, text='Company: ', bg='#fff').pack(side=LEFT)
Entry(company).pack(side=LEFT)

representative = LabelFrame(lesseeInfo)
representative.grid(sticky="W")
representative.configure(borderwidth=0)
Label(representative, text='Representative: ', bg='#fff').pack(side=LEFT)
Entry(representative).pack(side=LEFT)

location = LabelFrame(lesseeInfo)
location.grid(sticky="W")
location.configure(borderwidth=0)
Label(location, text='Location: ', bg='#fff').pack(side=LEFT)
Entry(location).pack(side=LEFT)

city = LabelFrame(lesseeInfo)
city.grid(sticky="W")
city.configure(borderwidth=0)
Label(city, text='City/State/ZIP: ', bg='#fff').pack(side=LEFT)
Entry(city).pack(side=LEFT)

phone = LabelFrame(lesseeInfo)
phone.grid(sticky="W")
phone.configure(borderwidth=0)
Label(phone, text='Phone: ', bg='#fff').pack(side=LEFT)
Entry(phone).pack(side=LEFT)




# SPACE
Label(root, text='', bg='#fff').grid(row=6)

# VEHICLE INFORMATION
Label(root, text='Vehicle Information', bg='#fff', font='Arial 15').grid(sticky='W', row=7, column=0)

# VEHICLE INFORMATION : COL 1
vehicleInfo1 = LabelFrame(root)
vehicleInfo1.grid(sticky="W", row=8, column=0)
vehicleInfo1.configure(borderwidth=0, bg='#fff')

VIN = LabelFrame(vehicleInfo1)
VIN.grid(sticky="W")
VIN.configure(borderwidth=0)
Label(VIN, text='VIN: ', bg='#fff').pack(side=LEFT)
Entry(VIN).pack(side=LEFT)

Make = LabelFrame(vehicleInfo1)
Make.grid(sticky="W")
Make.configure(borderwidth=0)
Label(Make, text='Make: ', bg='#fff').pack(side=LEFT)
Entry(Make).pack(side=LEFT)

Year = LabelFrame(vehicleInfo1)
Year.grid(sticky="W")
Year.configure(borderwidth=0)
Label(Year, text='Year: ', bg='#fff').pack(side=LEFT)
Entry(Year).pack(side=LEFT)

Color = LabelFrame(vehicleInfo1)
Color.grid(sticky="W")
Color.configure(borderwidth=0)
Label(Color, text='Color: ', bg='#fff').pack(side=LEFT)
Entry(Color).pack(side=LEFT)




# VEHICLE INFORMATION : COL 2
vehicleInfo2 = LabelFrame(root)
vehicleInfo2.grid(sticky="W", row=8, column=1)
vehicleInfo2.configure(borderwidth=0, bg='#fff')

regNo = LabelFrame(vehicleInfo2)
regNo.grid(sticky="W")
regNo.configure(borderwidth=0)
Label(regNo, text='Registration #: ', bg='#fff').pack(side=LEFT)
Entry(regNo).pack(side=LEFT)

model = LabelFrame(vehicleInfo2)
model.grid(sticky="W")
model.configure(borderwidth=0)
Label(model, text='Model: ', bg='#fff').pack(side=LEFT)
Entry(model).pack(side=LEFT)

mileage = LabelFrame(vehicleInfo2)
mileage.grid(sticky="W")
mileage.configure(borderwidth=0)
Label(mileage, text='Mileage: ', bg='#fff').pack(side=LEFT)
Entry(mileage).pack(side=LEFT)




# SPACE
Label(root, text='', bg='#fff').grid(row=9)




# TABLE
table = LabelFrame(root)
table.grid(sticky="W", row=10, column=0, columnspan=2)
table.configure(borderwidth=3, bg='#fff')


Label(table, text='VIN', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=0, column=0)
Label(table, text='Cost/Day', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=0, column=1)
Label(table, text='# of Days', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=0, column=2)
Label(table, text='Additional Costs', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=0, column=3)
Label(table, text='Line Total', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=0, column=4)

Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=1, column=0)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=1, column=1)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=1, column=2)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=1, column=3)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=1, column=4)

Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=2, column=0)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=2, column=1)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=2, column=2)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=2, column=3)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=2, column=4)

Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=3, column=0)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=3, column=1)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=3, column=2)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=3, column=3)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=3, column=4)



Label(table, text='Payment Method:', bg='#fff', borderwidth=0, width=46).grid(sticky="W", row=4, column=0, columnspan=3)

ch1 = LabelFrame(table)
ch1.grid(sticky="W", row=5, column=0, columnspan=3)
ch1.configure(borderwidth=0, width=46, bg='#fff')
cash = IntVar()
Checkbutton(ch1, text='Cash', variable=cash, bg='#fff').pack(side=LEFT)
checkNo = IntVar()
Checkbutton(ch1, text='Check No', variable=checkNo, bg='#fff').pack(side=LEFT)
Entry(ch1).pack(side=LEFT)

ch2 = LabelFrame(table)
ch2.grid(sticky="W", row=6, column=0, columnspan=3)
ch2.configure(borderwidth=0, width=46, bg='#fff')
creditNo = IntVar()
Checkbutton(ch2, text='Cash', variable=creditNo, bg='#fff').pack(side=LEFT)
Entry(ch2).pack(side=LEFT)

ch3 = LabelFrame(table)
ch3.grid(sticky="W", row=7, column=0, columnspan=3)
ch3.configure(borderwidth=0, width=46, bg='#fff')
creditNo = IntVar()
Checkbutton(ch3, text='Other', variable=creditNo, bg='#fff').pack(side=LEFT)
Entry(ch3).pack(side=LEFT)



Label(table, text='Subtotal:', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=4, column=3)
Label(table, text='Tax (     %):', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=5, column=3)
Label(table, text='Total:', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=6, column=3)
Label(table, text='Amount Paid:', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=7, column=3)

Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=4, column=4)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=5, column=4)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=6, column=4)
Label(table, text='', bg='#fff', borderwidth=1, width=15, relief="ridge").grid(row=7, column=4)


Label(root, text='', bg='#fff').grid()

sign = LabelFrame(root)
sign.grid()
sign.configure(borderwidth=0)
Label(sign, text='Authorized Signature: ', bg='#fff').pack(side=LEFT)
Entry(sign).pack(side=LEFT)

representativeName = LabelFrame(root)
representativeName.grid()
representativeName.configure(borderwidth=0)
Label(representativeName, text='Representative Name: ', bg='#fff').pack(side=LEFT)
Entry(representativeName).pack(side=LEFT)

btnApply = Button(root, text='Apply', command=submit).grid(sticky="W", row=23, column=1)


root.mainloop()
```
