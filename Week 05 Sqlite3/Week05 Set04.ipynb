# Week 5 Set 4: Connecting Tkinter with sqlite DB

**NOTE:** the database only stores values of the following fields: title, first name, last name, share with and business number

```python
from tkinter import *
import sqlite3

main = Tk()
main.title("Set 4")
main.config(bg="#fff")



#################
### DATABASES ###
#################

# create or connect to a db
conn = sqlite3.connect('./Database/set4.db')

# create a cursor
c = conn.cursor()

# create a table
# c.execute("""CREATE TABLE details (
#     title text,
#     fname text,
#     lname text,
#     share text,
#     number integer
# )""")

# create funmit function for database
def submit():
    conn = sqlite3.connect('./Database/set4.db')
    c = conn.cursor()
    
    # insert into table
    c.execute("INSERT INTO details VALUES (:title, :fname, :lname, :share, :number)", {
        'title': title.get(),
        'fname': fname.get(),
        'lname': lname.get(),
        'share': share.get(),
        'number': number.get()
    })
    
    c.execute("SELECT * FROM details")
    print('The values stored in the database are: ')
    print(c.fetchall())    
    
    conn.commit()
    conn.close()
    
    title.delete(0, END)
    fname.delete(0, END)
    lname.delete(0, END)
    share.delete(0, END)
    number.delete(0, END)
    
    





###############
### TKINTER ###
###############

root = LabelFrame(main, padx=10, pady=10)
root.pack(padx=30, pady=30)
root.configure(background="#fff", borderwidth=0)





###############
### TABLE 1 ###
###############

table1 = LabelFrame(root)
table1.grid(row=0, sticky='W')
table1.configure(background="#fff", borderwidth=0)


Label(table1, text="Tile", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=0, column=0)
title = Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60)
title.grid(row=0, column=1)

Label(table1, text="First name", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=1, column=0)
fname = Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60)
fname.grid(row=1, column=1)


Label(table1,text="Last name", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=2, column=0)
lname = Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60)
lname.grid(row=2, column=1)

Label(table1,text="Share with", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=3, column=0)
share = Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60)
share.grid(row=3, column=1)


Label(table1,text="Business number", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=4, column=0)
number = Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60)
number.grid(row=4, column=1)

Label(table1,text="Mobile number", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=5, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=5, column=1)

Label(table1,text="Email address", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=6, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=6, column=1)

Label(table1,text="Date of arrival", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=7, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=7, column=1)

Label(table1,text="Date of departure", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=8, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=8, column=1)

Label(table1,text="Name on credit card", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=9, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=9, column=1)

Label(table1,text="Credit card number", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=10, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=10, column=1)

Label(table1,text="Expiry date", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=11, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=11, column=1)

Label(table1,text="CVV number", bg='#fff', borderwidth=1, relief="solid", width=35).grid(row=12, column=0)
Entry(table1, bg='#fff', borderwidth=1, relief="solid", width=60).grid(row=12, column=1)

Label(table1, text="Payment method", bg='#fff', borderwidth=2, relief="solid", width=35).grid(row=13, column=0)
payment = LabelFrame(table1)
payment.grid(row=13, column=1, sticky='WE')
payment.configure(bg="#fff", borderwidth=1, relief="solid", width=60)
Label(payment, text="Credit card", bg='#fff', borderwidth=1, relief="solid").pack(side=LEFT, padx=20)
Label(payment,text="Direct bank transfer", bg='#fff', borderwidth=1, relief="solid").pack(side=LEFT)





Label(root, text='', bg='#fff').grid(row=1)

###############
### TABLE 2 ###
###############

table2 = LabelFrame(root)
table2.grid(row=2, sticky='W')
table2.configure(background="#fff", borderwidth=0)

Label(table2,text="----NEGOTIATED RATES----",font="Arial 10 bold", bg="#fff").grid(row=0, column=0, sticky='W')

Checkbutton(table2,text="Deluxe room single\tR1700", bg="#fff").grid(row=1, column=0, sticky='W')
Checkbutton(table2,text="Deluxe room double\tR1700", bg="#fff").grid(row=1, column=2, sticky='W')

Label(table2, text='\t\t', bg='#fff').grid(row=0, column=1)

Checkbutton(table2,text="Suites room single\tR1700", bg="#fff").grid(row=2, column=0, sticky='W')
Checkbutton(table2,text="Suites room double\tR1700", bg="#fff").grid(row=2, column=2, sticky='W')






Label(root, text='', bg='#fff').grid(row=3)

###############
### TABLE 3 ###
###############

table3 = LabelFrame(root)
table3.grid(row=4, sticky='W')
table3.configure(background="#fff", borderwidth=0)

Label(table3, text="----ROOM PREFERENCES----", font="Arial 10 bold", bg="#fff").grid(row=0, column=0, sticky='W')

Checkbutton(table3, text="king bed\t", bg="#fff").grid(row=1, column=0, sticky='W')
Checkbutton(table3, text="Twin - Two single beds", bg="#fff").grid(row=1, column=1, sticky='W')





Label(root, text='', bg='#fff').grid(row=5)

#############
### PARAS ###
#############

para = LabelFrame(root)
para.grid(row=6, sticky='W')
para.configure(background="#fff", borderwidth=0)

Label(para, text="The above rates are quoted per room, per night. The rates include breakfast, 14% vat, and Excludes 1% Tourism Levy and a voluntary\nR10 donation to the arabella community trust that will be levis onto your account\n", bg="#fff", justify=LEFT).grid(sticky='W')
Label(para, text="Total amount payable Hotel and Spa", bg="#fff").grid(sticky='W')
Label(para, text="ZAR_________x_________nights=ZAR_________due to Arabella", bg="#fff").grid(sticky='W')
Label(para, text="Credit card will be changed on receipt of this form and details will also be used to settle all incredentials not settle\non departure.A copy of the final folio will be sent to you should there be any unsettled changes", bg="#fff", justify=LEFT).grid(sticky='W')
Label(para, text="In order to qualify for the above rates, your booking needs to be made on or before 15th January 2016", bg="#fff").grid(sticky='W')
Label(para, text="Terms and conditions can be found on the next page", bg="#fff").grid(sticky='W')
Label(para, text="The rate is valid for seven days and after the conference dates.Check in time is 14:00 & check out time is 11:00", bg="#fff").grid(sticky='W')
Label(para, text="By your signature hereto,you are accepting all terms and conditions specified on this form and confirm that all information given is correct and accurate", bg="#fff").grid(sticky='W')






Label(root, text='', bg='#fff').grid(row=7)

##############
### FOOTER ###
##############

footer = LabelFrame(root)
footer.grid(row=8, sticky='W')
footer.configure(background="#fff", borderwidth=0)

Label(footer,text="Signature",bg="white").grid(row=0, column=0, sticky='W')
Entry(footer,width=30).grid(row=0, column=1)

Label(footer, text='     ', bg='#fff').grid(row=0, column=2)

Label(footer,text="Print Name",bg="white").grid(row=0,column=3, sticky='W')
Entry(footer,width=30).grid(row=0,column=4)

Label(footer,text="Date",bg="white").grid(row=1,column=0, sticky='W')
Entry(footer,width=30).grid(row=1,column=1)

submit=Button(root,text="Submit",fg="Black",bg="#eee", command=submit)
submit.grid(row=29,column=1,columnspan=3)

root.mainloop()
```
